```python
from pyspark.sql import SparkSession #mamy do czynienia z nowym sparkiem - bo data frame, mozliwosc wykonywania sql
# from pyspark import SparkContext --> tak by bylo gdyby byl stary spark
```

```python
spark = (
    SparkSession.builder
    .appName("Lab2-Transactions")
    .getOrCreate()
)
```

```python
spark
```

```python
spark.sparkContext.setLogLevel("WARN")
print(f"Spark {spark.version} — gotowy")
```

```python
df = spark.read.json("transactions_10k.jsonl")

print(f"Liczba rekordów: {df.count()}")
df.printSchema()
```

```python
df.show(10, truncate=False)
```

```python
from pyspark.sql.functions import to_timestamp, col

df = df.withColumn("timestamp", to_timestamp(col("timestamp"), "yyyy-MM-dd HH:mm:ss"))

df.printSchema()  # timestamp powinien być teraz 'timestamp (nullable = true)'
```

```python
# CZĘŚĆ 2
```

```python
#2.1
from pyspark.sql.functions import count, sum as _sum, avg, round as _round

store_summary = (
    df.groupBy("store")
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
        _round(avg("amount"), 2).alias("srednia_PLN"),
    )
    .orderBy("store")
)
store_summary.show()
```

```python
#2.2
from pyspark.sql.functions import min as _min, max as _max

store_summary = (
    df.groupBy("category")
    .agg(
        _min("amount").alias("MINIMUM"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
        _max("amount").alias("MAKSIMUM"),
    )
    .orderBy("category")
)
store_summary.show()
```

```python
#CZĘŚĆ 3
```

```python
#3.1
from pyspark.sql.functions import window

hourly = (
    df.groupBy(window("timestamp", "1 hour"))    # okno 1-godzinne
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
    )
    .orderBy("window")
)
hourly.show(truncate=False)
```

```python
hourly.printSchema()
```

```python
(
    hourly
    .select(
        col("window.start").alias("od"),
        col("window.end").alias("do"),
        "liczba_tx",
        "suma_PLN",
    )
    .show(truncate=False)
)
```

```python
#ZADANIE 3.2

from pyspark.sql.functions import window

hourly = (
    df.groupBy(window("timestamp", "30 minutes"), "store")    # okno 30 min
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
    )
    .orderBy("window", "store")
)
hourly.show(truncate=False)
```

```python
#ZADANIE 3.3
from pyspark.sql.functions import desc

hourly = (
    df.groupBy(window("timestamp", "1 hour"), "store")    
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
    )
    .filter(col("store") == "Kraków")  
    .orderBy(desc("suma_PLN"))          
)
hourly.show(truncate=False)
```

odp: “Kraków” miał najwyższy przychód od 9 do 10

```python
# 4.1
sliding = (
    df.groupBy(window("timestamp", "1 hour", "30 minutes"))  # szerokość 1h, krok 30min
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
    )
    .select(
        col("window.start").alias("od"),
        col("window.end").alias("do"),
        "liczba_tx",
        "suma_PLN",
    )
    .orderBy("od")
)
sliding.show(truncate=False)
```

```python
tumbling_rows = (
    df.groupBy(window("timestamp", "1 hour"))
    .agg(count("tx_id"))
    .count()
)
sliding_rows = (
    df.groupBy(window("timestamp", "1 hour", "30 minutes"))
    .agg(count("tx_id"))
    .count()
)
print(f"Tumbling (1h):          {tumbling_rows} okien")
print(f"Sliding  (1h / 30min):  {sliding_rows} okien")
```

Sliding daje więcej wierszy, ponieważ są tej samej długości, ale się zachodzą, więc więcej ich się zmieści.

```python
# 1.

hourly = (
    df.groupBy(window("timestamp", "1 hour"), "store")   
    .agg(
        _round(avg("amount"), 2).alias("suma_PLN"),
    )
    .filter(col("store") == "Gdańsk")  
    .orderBy("suma_PLN")          
)
hourly.show(truncate=False)
```

```python
# 2.
from pyspark.sql.functions import lit
hourly = (
    df.groupBy(window("timestamp", "30 minutes"), "category")   
    .agg(
        count("tx_id").alias("liczba_tx"),
    )
    .filter(col("window").end == lit("2026-04-12 09:30:00"))         
)
hourly.show(truncate=False)
```

```python
# 3.
# 2.
from pyspark.sql.functions import lit
hourly = (
    df.groupBy(window("timestamp", "15 minutes"))   
    .agg(
        count("tx_id").alias("liczba_tx"),
    )
    .orderBy(desc("liczba_tx"))       
)
hourly.show(truncate=False)
```
